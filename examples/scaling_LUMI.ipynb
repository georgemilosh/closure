```python
import matplotlib.pyplot as plt
import pandas as pd
import re



# Function to extract and multiply integers from the specified format
def extract_number(name):
    match = re.search(r'-(\d+)\.out$', name)
    if match:
        return int(match.group(1))
    match = re.search(r'out(\d+)\.out$', name)
    if match:
        return int(match.group(1))
    return None

# Read the CSV file
df = []

df = (pd.read_csv(f'/volume1/scratch/share_dir/ecsim/brecht/strong_scaling.csv'))
df = df.dropna()
df = df.reset_index(drop=True)
# Apply the function to the 'name' column
df['cores'] = df['name'].apply(extract_number)
df['speedup'] = df['Actual'][0] / df['Actual']
df['efficiency'] = df['cores'][0]* df['speedup'] / df['cores']
df
```

```python
fig, axes = plt.subplots(1, 2, figsize=(8, 4))

# Plot on the left panel
axes[0].scatter(df['cores'], df['efficiency'])
axes[0].set_xscale('log')
axes[0].set_title('Efficiency')
axes[0].set_xlabel('Cores')
axes[0].set_ylabel('Efficiency')

# Plot on the right panel
axes[1].scatter(df['cores'], df['speedup'])
axes[1].set_xscale('log')
axes[1].set_title('Speedup')
axes[1].set_xlabel('Cores')
axes[1].set_yscale('log')
axes[1].set_ylabel('Speedup')
#axes[1].plot([df_mean['cores'].min(), df_mean['cores'].max()], [df_mean['cores'].min(), df_mean['cores'].max()], color='red', linestyle='--')

plt.tight_layout()
plt.show()
```
